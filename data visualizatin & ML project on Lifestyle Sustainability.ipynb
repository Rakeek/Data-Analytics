## Feature Selection

### Logistic Regression

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,precision_score,classification_report,confusion_matrix
```

```python
df.columns
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
for column in['DietType','LocalFoodFrequency','TransportationMode','UsingPlasticProducts','DisposalMethods','PhysicalActivities']: # instead of writing many lines for each columns we can use a for loop to do it in one line.
    df[column] = le.fit_transform(df[column])
```

```python
df
```

```python
x=df[['DietType','LocalFoodFrequency','TransportationMode','UsingPlasticProducts','DisposalMethods','PhysicalActivities']]
x
```

```python
y=df[['Rating']]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train,x_test,y_train,y_test)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model = LogisticRegression()
model.fit(x_train,y_train)
model
```

```python
y_predict=model.predict(x_test)
y_predict
```

```python
y_test
```

```python
cnf = confusion_matrix(y_test,y_predict)
print(cnf)
```

```python
log_acc = accuracy_score(y_test,y_predict)
print(log_acc)
```

```python
print(classification_report(y_test,y_predict)) # summary of your model performance.
```

```python
# predict the new data 
new_experience=np.array([[0,3,2,3,3,2]])

predicted_rating=model.predict(new_experience)
print(f'predicted Sustainable Lifestyle Rating (0-5): {predicted_rating[0]}')
```

## KNN

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,precision_score,classification_report,confusion_matrix
```

```python
x=df[['DietType','LocalFoodFrequency','TransportationMode','UsingPlasticProducts','DisposalMethods','PhysicalActivities']]
x
```

```python
y=df[['Rating']]
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
print(x_train,x_test,y_train,y_test)
```

```python
knn = KNeighborsClassifier(n_neighbors=5) # we are giving one numeric value to k = 5
knn.fit(x_train,y_train)
knn
```

```python
y_pred=knn.predict(x_test)
y_pred
```

```python
y_test
```

```python
cnf = confusion_matrix(y_test,y_pred)
print(cnf)
```

```python
knn_acc = accuracy_score(y_test,y_pred)
print(knn_acc)
```

```python
print(classification_report(y_test,y_pred)) # summary of your model performance.
```

```python
# predict the new data 
new_experiencee=np.array([[2,1,0,2,1,0]])

predicted_rating=model.predict(new_experiencee)
print(f'predicted Sustainable Lifestyle Rating (0-5): {predicted_rating[0]}')
```

```python
models = ['LogisticRegression', 'KNeighborsClassifier']
accuracies = [log_acc,knn_acc,]
plt.figure(figsize=(8, 6))
plt.bar(models, accuracies, color=['skyblue', 'grey', 'brown'])
plt.title('Model Comparison: Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Models')
plt.show()
```
